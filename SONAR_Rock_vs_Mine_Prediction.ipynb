# SONAR Rock vs Mine Prediction
End To End Python Machine Learning

## Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data collection and data processing

```python
#loading the dataset
data = pd.read_csv("/content/Sonar data.csv", header = None)
print(data)
```

```python
#first five values
data.head()
```

```python
#last five values
data.tail()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
#no. of rows and columns
data.shape
```

```python
data.describe()
```

```python
#R(Rock) and M(Mine) are mentioned in 60th column
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

## separating data and labels


```python
X = data.drop(columns = 60, axis = 1)
Y = data[60]
```

```python
print(X)
```

```python
print(Y)
```

## Training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Trainig the model

(Logistic Regression)

```python
model = LogisticRegression()
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
#training the Logistic regression model using the training data
model.fit(X_train, Y_train)
```

## Evaluating the model

```python
#finding accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
#finding accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```
